## list comprehension

```python
# Add 1 to 10 numbers in a list
#without list comprehension
l1=[]
for i in range(1,11):
    l1.append(i)
print(l1)
#with list comprehension
l=[i for i in range (1,11)]
l
```

```python
# scalar multiplication of a vector
v=[2,3,4]
s=-3
#without list comprehension
x=[]
for i in v:
    x.append(i*s)
print(x)

#with list comprehension
[i*s for i in v]    
```

```python
#Add squares
l=[1,2,3,4,5]

[i**2 for i in l]
```

```python
# print all the numbers divisible by 5 in the range of 1 to 5
[i for i in range (1,51,1) if i%5==0]
```

```python
# find languages which start with letter p
languages=['java','python','php','c','javascript']
[language for language in languages if language.startswith('p')]
```

```python
# nested if with list comprehension
basket=['apple','guave','cherry','banana']
my_fruits=['apple','kiwi','grapes','banana']
# add new list form my_fruits and items if the fruit exist in basket and also start with 'a'
[fruit for fruit in my_fruits if fruit in basket if fruit.startswith('a')]
```

```python
#print a (3,3) using list comprehension --> nested list comprehension
[[i*j for i in range (1,4)]for j in range(1,4)]
```

```python
# cartesian product
l1=[1,2,3,4]
l2=[5,6,7,8]
[i*j for i in l1  for j in l2]
```

## traverse a list by 
* itemwise
* indexwise

```python
#itemwise
l=[1,2,3,4]
for i in l:
    print(i)
```

```python
#indexwise
l=[1,2,3,4]
for i in range(0,len(l)):
    print(i)
```
