## Creating Tuples

```python
# empty tuples
t=()
print(t)
# create a tuple with a single item
t1=(2,)
print(type(t1))
print(t1)
#homogenous tuple
t3=(1,2,3,4)
print(t3)
#heterogenous tuple
t4=(1,2.5,True,[1,2,3])
print(t4)
print(type(t4))
# using type conversion 
t5=tuple('hello')
print(t5)
```

## Accessing items
* indexing
* slicing

```python
print(t3)
print(t3[0])
print(t3[-1])
print(t3[2:4]) # 3 4
print(t3[:3:2]) # 1 3
print(t3[1::]) # 2 3 4
print(t3[::1]) # 1 2 3 4
print(t3[::-1]) # 4 3 2 1  
```

## editing items

```python
print(t3)
t3[0]=100 # immutable just like string
```

## Adding items

```python
print(t3)
#not possible
```

## Deleting items

```python
print(t3)
del t3 # deletion only work when you delete complete tuple otherwise cause error 
print(t3)
```

## Operation on tuples
* arthimetric(+,*)
* loop
* membership

```python
t1=(1,2,3,4)
t2=(5,6,7,8)

print(t1+t2)

print(t1*3)

#membership
1 in t1 

for i in t1:
    print(i)
```

## Tuple Functions

```python
# len/sum/max/min/sorted/
t1=(1,2,5,3,4)
print(len(t1))

print(min(t1))
print(max(t1))
print(sum(t1))
print(sorted(t1))
```

```python
#count
t5=(1,2,3,1,2,3,1,2,3,11,2,3,3,42,11,2,3,21,)
print(t5.count(2))
print(t5.count(3))
```

```python
#index
t5.index(3)
```

## special syntax

```python
#tuple unpacking
a,b,c=(1,2,3)
print(a,b,c)
```

```python
a,b=(1,2,3)
```

```python
a=1
b=2
a,b=b,a
print(a,b)
```

```python
a,b,*other=(1,2,3,4)
print(a,b)
print(*other)
```

```python
#zipping tuple
t1=(1,2,3,4)
t2=(5,6,7,8)

tuple(zip(t1,t2))
```

## SET DATA STRUCTURE

```python
#empty set
s=set()
print(s)
print(type(s))
#1d set
s1={1,2,3}
print(s1)
#2D Set
#s2={1,2,3,{4,5}} # give an error because we cant create two sets in single set

#heterogenous
s3={1,4.5,'hello',(1,2,3)}
print(s3)

#using type conversion
s4 =set([1,2,3,4])
print(s4)

s5 =set((1,2,3))
print(s5)

#duplication is not allowed
s6={1,1,2,3,3,2}
print(s6)

#unordered
s1={1,2,3}
s2={3,2,1}

print(s1 == s2)
```

#### Accessing item & Editing items are not allowed in set data structure
